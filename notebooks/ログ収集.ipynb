```python
import requests
from bs4 import BeautifulSoup
import os

path=r"http://133.167.32.100/aiwolf/2025/JSAI/MainTruck5/Game/"#ここにURL
response = requests.get(path)
soup = BeautifulSoup(response.text, 'html.parser')

log_links = [a['href'] for a in soup.find_all('a', href=True) if a['href'].endswith('.log')]

os.makedirs('2024_spring', exist_ok=True)

for link in log_links:
    log_url = path + link
    log_response = requests.get(log_url)
    with open(os.path.join('2024_spring', link), 'wb') as f:
        f.write(log_response.content)
```

```python
import glob
log_files = glob.glob('2024_spring/*.log')
len(log_files)
```

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
result=[]
for log_file in log_files:
    try:
        df=pd.read_csv(log_file, names=["date","type","info1","info2","info3","info4"],index_col=False)
        df=df[df["type"].isin(["result","status"])]
        df=df.drop_duplicates(subset=["type"],keep="last")
        result.append(df[df["type"]=="result"]["info3"].values[0])
    except:
        result.append("error")
result_df=pd.DataFrame(result,columns=["result"])
result_df["result"].value_counts()
```

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
pass_list=[]
for log_file in log_files:
    try:
        df=pd.read_csv(log_file, names=["date","type","info1","info2","info3","info4","info5"],index_col=False)
        df=df[df["type"].isin(["result","status"])]
        df=df.drop_duplicates(subset=["type"],keep="last")
        result=df[df["type"]=="result"]["info3"].values[0]
        #print(result)
        if result!="NONE":
            pass_list.append(log_file)
    except:
        continue
len(pass_list)
```

```python
import shutil
import os
os.makedirs('2024_spring_pass', exist_ok=True)
for log_file in pass_list:
    shutil.copy(log_file, '2024_spring_pass')
```
